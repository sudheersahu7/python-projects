```python
import csv
while True:
    student_name = input ("student name:")
    student_age = input ("age of student:")
    course_name = input ("which course you enroll:")
    with open ("student.csv" , "a" , newline="") as file:
        writer = csv.writer(file)
        writer.writerow ([student_name , student_age , course_name])
    print ("data saved")
    stop = input("add another student data ? y/n")
    if stop.lower() == "n":
        break
        
        
        
    
```
